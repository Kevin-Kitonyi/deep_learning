```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn
%matplotlib inline
```

```python
my_df = pd.DataFrame(randn(100, 4), columns=["Mon", "Tue", "Wed", "Thur"])
my_df
```

```python
plt.hist(my_df["Wed"])
plt.show()
```

```python
plt.hist(my_df["Wed"], bins = 30)
plt.show()
```

```python
plt.plot(my_df)
plt.show()
```

```python
my_df["Wed"].plot(kind = 'hist', bins=30)
plt.show()
```

```python
my_df.abs().plot(kind = 'area',  stacked = False)
plt.show()
```

```python
my_df.abs().plot(kind = 'area', alpha=0.4, title="My Awesome area plot")
plt.show()
```

```python
my_df.abs().plot(kind = 'area', alpha=0.4, title="My Awesome area plot", legend=False)
plt.show()
```

```python
my_df["Wed"].abs().plot(kind = 'area', alpha=0.4, title="My Awesome area plot", legend=False)
plt.show()
```
